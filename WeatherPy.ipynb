```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

#print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city_in = 0
city_array = []
units = "imperial"
for city in cities:
    try:
        
        url = "http://api.openweathermap.org/data/2.5/weather?q=" + city + "&appid=" + weather_api_key + "&units=" + units
     #url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid{weather_api_key}&units={units}"
        response = requests.get(url).json()
    #results=response#.json()
        #pprint(response) ***Response prints out. This line is good. 
               
        city_in +=1
        date = (response['dt'])
        city_name = (response['name'])
        humidity = (response['main']['humidity'])
        longitude = (response['coord']['lon'])
        latitude = (response['coord']['lat'])
        country = (response['sys']['country'])
        cloudiness = (response['clouds']['all'])
        max_temp = (response['main']['temp_max'])
        wind_speed = (response['wind']['speed'])
       
    
        city_array.append({"Date": date, 
                           "City": city_name,
                           "Country": country, 
                           "Lat": latitude, 
                           "Long": longitude, 
                           "Humid": humidity, 
                           "Max Temp": max_temp, 
                           "Wind": wind_speed, 
                           "Cloudy": cloudiness})
        
        print(f"Please Work: {city_in} of {len(cities)} |{city}")
              
    except: 
        
        pass
    #indspeed = results [""]
print(city_array)
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def double(x):
    return 2*x
```

```python
# Create Northern and Southern Hemisphere DataFrames
north = weather_city.loc[weather_city["Lat"] >=0]
south = weather_city.loc[weather_city["Lat"] <=0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north["Lat"]
y_values = north["Max Temp"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)#stack overflow and homework

print(f"r^2 is: {rvalue**2}")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

regress_values = x_values * slope + intercept
plt.plot(x_values,regress_values,"r-")


plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.title("N Hemi Max Temp vs Latitude Linear Regreassion")
plt.scatter(north["Lat"], north["Max Temp"], facecolors="blue", marker="+", edgecolor="black")
plt.grid(True)
plt.savefig("../output_data/N Hemi vs Latitude Linear.png")
plt.show()
```
